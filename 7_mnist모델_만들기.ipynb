```python
from ultralytics import YOLO

number_model = YOLO('yolov8n.pt')
data_path = "numbers/data.yaml"
number_model.train(data=data_path, epochs=20, imgsz=640)
metrics = number_model.val(data=data_path)
metrics
```

```python
from ultralytics import YOLO
tune_number_model = YOLO('yolov8n.pt')

tune_number_model.tune(
    data='numbers/data.yaml',
    epochs=15,
    imgsz=640,
    iterations=15,
    val=True
)
```

```python
import yaml  
data_path = 'numbers/data.yaml'
with open("./runs/detect/tune3/best_hyperparameters.yaml", "r", encoding="utf-8") as f:
    hyp_dict = yaml.safe_load(f)
    
tune_number_model.train(data=data_path, epochs=80, imgsz=640, batch=16, **hyp_dict)
```
